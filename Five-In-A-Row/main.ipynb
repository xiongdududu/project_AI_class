```python
import sys
import random
import pygame
from pygame.locals import *
import pygame.gfxdraw
from checkerboard import Check_all, Black, White, vector, Position


Length = 30  # distance between nodes
Node_Numbers = 15  # number of nodes on each row/column
Outer_Width = 20  # outer width
Border_Width = 4  # border width
Inside_Width = 4  # distance between border and outer
Border_Length = Length * (Node_Numbers - 1) + Inside_Width * 2 + Border_Width  # border length
Start_X = Start_Y = Outer_Width + int(Border_Width / 2) + Inside_Width  # original point
SCREEN_HEIGHT = Length * (Node_Numbers - 1) + Outer_Width * 2 + Border_Width + Inside_Width * 2  # screen height
SCREEN_WIDTH = SCREEN_HEIGHT + 200  # screen width

Radius = Length // 2 - 3  # radius of chess
Radius2 = Length // 2 + 3 # radius of chess sample
Board_Color = (180,150, 60) # color of border (brown)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (200, 30, 30)
BRIGHT_RED=(255,30,30)
BLUE = (30, 30, 200)
GREEN=(0,200,0)
BRIGHT_GREEN = (0,255,0)

RIGHT_INFO_POS_X = SCREEN_HEIGHT + Radius2 * 2 + 10

clock = pygame.time.Clock()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
def game_intro():
    pygame.init()
    intro = True
    screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    while intro:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
        screen.fill(WHITE)
        button("level_1", 150, 190, 90, 40, GREEN, BRIGHT_GREEN,d1)
        button("level_2", 300, 190, 90, 40, GREEN, BRIGHT_GREEN,d2)
        button("level_3", 450, 190, 90, 40, GREEN, BRIGHT_GREEN,d3)
        pygame.display.update()
        clock.tick(15)
    
def _text(screen, font, x, y, text, color=(255, 255, 255)):
    imgText = font.render(text, True, color)
    screen.blit(imgText, (x, y))    
    
def d1():

    pygame.display.set_caption('five in a row')

    font1 = pygame.font.SysFont('SimHei', 32)
    font2 = pygame.font.SysFont('SimHei', 72)
    fwidth, fheight = font2.size('Black Win')

    checkerboard = Check_all(Node_Numbers)
    current_runner = Black
    winner = None
    com = computer_step(Node_Numbers, White)

    black_win_count = 0
    white_win_count = 0




    while True :
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                quit()
            elif event.type == KEYDOWN:
                if event.key == K_RETURN:
                    if winner is not None:
                        winner = None
                        current_runner = Black
                        checkerboard = Check_all(Node_Numbers)
                        com = computer_step(Node_Numbers, White)
            elif event.type == MOUSEBUTTONDOWN:
                if winner is None:
                    pressed_array = pygame.mouse.get_pressed()
                    if pressed_array[0]:
                        mouse_pos = pygame.mouse.get_pos()
                        click_point = _get_clickposition(mouse_pos)
                        if click_point is not None:
                            if checkerboard.able_to_drop(click_point):
                                winner = checkerboard.step(current_runner, click_point)
                                if winner is None:
                                    current_runner = _get_next(current_runner)
                                    com.get_people_drop(click_point)
                                    computer_step_point = com.computer_drop1()
                                    winner = checkerboard.step(current_runner, computer_step_point)
                                    if winner is not None:
                                        white_win_count += 1
                                    current_runner = _get_next(current_runner)
                                else:
                                    black_win_count += 1
                        else:
                            print('out of board')
        button("back", SCREEN_HEIGHT + Radius2*2, SCREEN_HEIGHT - Radius2 * 13, 90, 40, RED, BRIGHT_RED,game_intro)
        pygame.display.update()
        clock.tick(50)
        # draw board
        _draw_board(screen)

        # draw chess already dropped
        for i, row in enumerate(checkerboard.checkerboard):
            for j, cell in enumerate(row):
                if cell == Black.Value:
                    _draw_chess(screen, Position(j, i), Black.Color)
                elif cell == White.Value:
                    _draw_chess(screen, Position(j, i), White.Color)

        _draw_left_info(screen, font1, current_runner, black_win_count, white_win_count)

        if winner:
            _text(screen, font2, (SCREEN_WIDTH - fwidth)//2, (SCREEN_HEIGHT - fheight)//2, winner.Name + ' win', BRIGHT_RED)

        pygame.display.flip()
def d2():

    pygame.display.set_caption('five in a row')

    font1 = pygame.font.SysFont('SimHei', 32)
    font2 = pygame.font.SysFont('SimHei', 72)
    fwidth, fheight = font2.size('Black Win')

    checkerboard = Check_all(Node_Numbers)
    current_runner = Black
    winner = None
    com = computer_step(Node_Numbers, White)

    black_win_count = 0
    white_win_count = 0


    while True :
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                quit()
            elif event.type == KEYDOWN:
                if event.key == K_RETURN:
                    if winner is not None:
                        winner = None
                        current_runner = Black
                        checkerboard = Check_all(Node_Numbers)
                        com = computer_step(Node_Numbers, White)
            elif event.type == MOUSEBUTTONDOWN:
                if winner is None:
                    pressed_array = pygame.mouse.get_pressed()
                    if pressed_array[0]:
                        mouse_pos = pygame.mouse.get_pos()
                        click_point = _get_clickposition(mouse_pos)
                        if click_point is not None:
                            if checkerboard.able_to_drop(click_point):
                                winner = checkerboard.step(current_runner, click_point)
                                if winner is None:
                                    current_runner = _get_next(current_runner)
                                    com.get_people_drop(click_point)
                                    computer_step_point = com.computer_drop2()
                                    winner = checkerboard.step(current_runner, computer_step_point)
                                    if winner is not None:
                                        white_win_count += 1
                                    current_runner = _get_next(current_runner)
                                else:
                                    black_win_count += 1
                        else:
                            print('out of board')

                            
        button("back", SCREEN_HEIGHT + Radius2*2, SCREEN_HEIGHT - Radius2 * 13, 90, 40, RED, BRIGHT_RED,game_intro)
        pygame.display.update()
        clock.tick(50)                      
        
        # draw board
        _draw_board(screen)

        # draw chess already dropped
        for i, row in enumerate(checkerboard.checkerboard):
            for j, cell in enumerate(row):
                if cell == Black.Value:
                    _draw_chess(screen, Position(j, i), Black.Color)
                elif cell == White.Value:
                    _draw_chess(screen, Position(j, i), White.Color)

        _draw_left_info(screen, font1, current_runner, black_win_count, white_win_count)

        if winner:
            _text(screen, font2, (SCREEN_WIDTH - fwidth)//2, (SCREEN_HEIGHT - fheight)//2, winner.Name + ' win', BRIGHT_RED)

        pygame.display.flip()
        
def d3():

    pygame.display.set_caption('five in a row')

    font1 = pygame.font.SysFont('SimHei', 32)
    font2 = pygame.font.SysFont('SimHei', 72)
    fwidth, fheight = font2.size('Black Win')

    checkerboard = Check_all(Node_Numbers)
    current_runner = Black
    winner = None
    com = computer_step(Node_Numbers, White)

    black_win_count = 0
    white_win_count = 0


    while True :
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                quit()
            elif event.type == KEYDOWN:
                if event.key == K_RETURN:
                    if winner is not None:
                        winner = None
                        current_runner = Black
                        checkerboard = Check_all(Node_Numbers)
                        com = computer_step(Node_Numbers, White)
            elif event.type == MOUSEBUTTONDOWN:
                if winner is None:
                    pressed_array = pygame.mouse.get_pressed()
                    if pressed_array[0]:
                        mouse_pos = pygame.mouse.get_pos()
                        click_point = _get_clickposition(mouse_pos)
                        if click_point is not None:
                            if checkerboard.able_to_drop(click_point):
                                winner = checkerboard.step(current_runner, click_point)
                                if winner is None:
                                    current_runner = _get_next(current_runner)
                                    com.get_people_drop(click_point)
                                    computer_step_point = com.computer_drop3()
                                    winner = checkerboard.step(current_runner, computer_step_point)
                                    if winner is not None:
                                        white_win_count += 1
                                    current_runner = _get_next(current_runner)
                                else:
                                    black_win_count += 1
                        else:
                            print('out of board')

                            
        button("back", SCREEN_HEIGHT + Radius2*2, SCREEN_HEIGHT - Radius2 * 13, 90, 40, RED, BRIGHT_RED,game_intro)
        pygame.display.update()
        clock.tick(50)                      
        
        # draw board
        _draw_board(screen)

        # draw chess already dropped
        for i, row in enumerate(checkerboard.checkerboard):
            for j, cell in enumerate(row):
                if cell == Black.Value:
                    _draw_chess(screen, Position(j, i), Black.Color)
                elif cell == White.Value:
                    _draw_chess(screen, Position(j, i), White.Color)

        _draw_left_info(screen, font1, current_runner, black_win_count, white_win_count)

        if winner:
            _text(screen, font2, (SCREEN_WIDTH - fwidth)//2, (SCREEN_HEIGHT - fheight)//2, winner.Name + ' win', BRIGHT_RED)

        pygame.display.flip()

def text_objects(text, font):
    textSurface = font.render(text, True, BLACK)
    return textSurface, textSurface.get_rect()

def button (msg, x, y, w, h, ic, ac, action=None):

    mouse =pygame.mouse.get_pos()
    click = pygame.mouse.get_pressed()
#     print(click)
    if x + w > mouse[0] > x and y + h > mouse[1] > y:
        pygame.draw.rect(screen, ac, (x,y,w,h))
        if click[0] == 1 and action != None:
            action()
    else:
        pygame.draw.rect(screen, ic, (x,y,w,h))
    smallText = pygame.font.SysFont('comicsansms', 20)
    textSurf, textRect = text_objects(msg, smallText)
    textRect.center = ( (x+(w/2)), (y+(h/2)))
    screen.blit(textSurf, textRect)
        
def _get_next(current_runner):
    if current_runner == Black:
        return White
    else:
        return Black
    
# draw board
def _draw_board(screen):
    # background color
    screen.fill(Board_Color)
    # draw border
    pygame.draw.rect(screen, BLACK, (Outer_Width, Outer_Width, Border_Length, Border_Length), Border_Width)
    # draw lines
    for i in range(Node_Numbers):
        pygame.draw.line(screen, BLACK,
                         (Start_Y, Start_Y + Length * i),
                         (Start_Y + Length * (Node_Numbers - 1), Start_Y + Length * i),
                         1)
    for j in range(Node_Numbers):
        pygame.draw.line(screen, BLACK,
                         (Start_X + Length * j, Start_X),
                         (Start_X + Length * j, Start_X + Length * (Node_Numbers - 1)),
                         1)
            
# draw chess
def _draw_chess(screen, position, chess_color):
    # pygame.draw.circle(screen, chess_color, (Start_X + Length * position.X, Start_Y + Length * position.Y), Stone_Radius)
    pygame.gfxdraw.aacircle(screen, Start_X + Length * position.X, Start_Y + Length * position.Y, Radius, chess_color)
    pygame.gfxdraw.filled_circle(screen, Start_X + Length * position.X, Start_Y + Length * position.Y, Radius, chess_color)
    
# draw info
def _draw_left_info(screen, font, current_runner, black_win_count, white_win_count):
    _draw_chess_pos(screen, (SCREEN_HEIGHT + Radius2, Start_X + Radius2), Black.Color)
    _draw_chess_pos(screen, (SCREEN_HEIGHT + Radius2, Start_X + Radius2 * 4), White.Color)

    _text(screen, font, RIGHT_INFO_POS_X, Start_X + 3, 'player', BLUE)
    _text(screen, font, RIGHT_INFO_POS_X, Start_X + Radius2 * 3 + 3, 'computer', BLUE)

    _text(screen, font, SCREEN_HEIGHT, SCREEN_HEIGHT - Radius2 * 8, 'history：', BLUE)
    _draw_chess_pos(screen, (SCREEN_HEIGHT + Radius2, SCREEN_HEIGHT - int(Radius2 * 4.5)), Black.Color)
    _draw_chess_pos(screen, (SCREEN_HEIGHT + Radius2, SCREEN_HEIGHT - Radius2 * 2), White.Color)
    _text(screen, font, RIGHT_INFO_POS_X, SCREEN_HEIGHT - int(Radius2 * 5.5) + 3, f'{black_win_count} win', BLUE)
    _text(screen, font, RIGHT_INFO_POS_X, SCREEN_HEIGHT - Radius2 * 3 + 3, f'{white_win_count} win', BLUE)
    

def _draw_chess_pos(screen, pos, chess_color):
    pygame.gfxdraw.aacircle(screen, pos[0], pos[1], Radius2, chess_color)
    pygame.gfxdraw.filled_circle(screen, pos[0], pos[1], Radius2, chess_color)
    
# mouse position(x,y)
def _get_clickposition(click_pos):
    pos_x = click_pos[0] - Start_X
    pos_y = click_pos[1] - Start_Y
    if pos_x < -Inside_Width or pos_y < -Inside_Width:
        return None
    x = pos_x // Length
    y = pos_y // Length
    if pos_x % Length > Radius:
        x += 1
    if pos_y % Length > Radius:
        y += 1
    if x >= Node_Numbers or y >= Node_Numbers:
        return None

    return Position(x, y)


class computer_step:
    def __init__(self, node_numbers, chess):
        self._node_numbers = node_numbers
        self._computer = chess
        if chess == White:
            self._people = Black
        else:
            self._people = White
        self._board = [[0] * node_numbers for _ in range(node_numbers)]

    def get_people_drop(self, position):
        self._board[position.Y][position.X] = self._people.Value

    def computer_drop1(self):
        position = None
        score = 0
        for i in range(self._node_numbers):
            for j in range(self._node_numbers):
                if self._board[j][i] == 0:
                    score1 = self._get_position_score1(Position(i, j))
                    if score1 > score:
                        score = score1
                        position = Position(i, j)
                    elif score1 == score and score1 > 0:
                        u = random.randint(0, 1)
                        if u  == 0:
                            position = Position(i, j)
        self._board[position.Y][position.X] = self._computer.Value
        return position

    def _get_position_score1(self, position):
        score = 0
        for v in vector:
            score += self._get_vector_score1(position, v[0], v[1])
        return score

    def _get_vector_score1(self, position, x_vector, y_vector):
        count = 0   # number of computer continuous chess
        count1 = 0  # number of people continuous chess
        blank = None   # blank between continuous chess of computer
        blank1 = None  # blank between continuous chess of people
        sides = 0    # block on continuous chess (two sides) of computer
        sides1 = 0   # block on continuous chess (two sides) of people

        # 1 means computer chess besides, 2 means people chess
        temp = self._get_chess_color(position, x_vector, y_vector, True)
        if temp != 0:
            for i in range(1, 6):
                x = position.X + i * x_vector
                y = position.Y + i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # blank = 2, quit
                    elif temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # blocked by opponent
                    if temp == 1:
                        sides += 1
                    elif temp == 2:
                        sides1 += 1

        if blank is False:
            blank = None
        if blank1 is False:
            blank1 = None

        _temp = self._get_chess_color(position, -x_vector, -y_vector, True)
        if _temp != 0:
            for i in range(1, 6):
                x = position.X - i * x_vector
                y = position.Y - i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if _temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # blank = 2, quit
                    elif _temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # blocked by opponent
                    if _temp == 1:
                        sides += 1
                    elif _temp == 2:
                        sides1 += 1

        score = 0
        if count == 4:
            score = 10000
        elif count1 == 4:
            score = 9000
        elif count == 3:
            if sides == 0:
                score = 1000
            elif sides == 1:
                score = 100
            else:
                score = 0
        elif count1 == 3:
            if sides1 == 0:
                score = 1
            elif sides1 == 1:
                score = 1
            else:
                score = 0
        elif count == 2:
            if sides == 0:
                score = 100
            elif sides == 1:
                score = 10
            else:
                score = 0
        elif count1 == 2:
            if sides1 == 0:
                score = 1
            elif sides1 == 1:
                score = 1
            else:
                score = 0
        elif count == 1:
            score = 10
        elif count1 == 1:
            score = 1
        else:
            score = 0

        if blank or blank1:
            score /= 2

        return score
    
    
    def computer_drop2(self):
        position = None
        score = 0
        for i in range(self._node_numbers):
            for j in range(self._node_numbers):
                if self._board[j][i] == 0:
                    score1 = self._get_position_score2(Position(i, j))
                    if score1 > score:
                        score = score1
                        position = Position(i, j)
                    elif score1 == score and score1 > 0:
                        u = random.randint(0, 1)
                        if u  == 0:
                            position = Position(i, j)
        self._board[position.Y][position.X] = self._computer.Value
        return position

    def _get_position_score2(self, position):
        score = 0
        for v in vector:
            score += self._get_vector_score2(position, v[0], v[1])
        return score

    def _get_vector_score2(self, position, x_vector, y_vector):
        count = 0   # number of computer continuous chess
        count1 = 0  # number of people continuous chess
        blank = None   # blank between continuous chess of computer
        blank1 = None  # blank between continuous chess of people
        sides = 0    # block on continuous chess (two sides) of computer
        sides1 = 0   # block on continuous chess (two sides) of people

        # 1 means computer chess besides, 2 means people chess
        temp = self._get_chess_color(position, x_vector, y_vector, True)
        if temp != 0:
            for i in range(1, 6):
                x = position.X + i * x_vector
                y = position.Y + i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # blank = 2, quit
                    elif temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # block by opponent
                    if temp == 1:
                        sides += 1
                    elif temp == 2:
                        sides1 += 1

        if blank is False:
            blank = None
        if blank1 is False:
            blank1 = None

        _temp = self._get_chess_color(position, -x_vector, -y_vector, True)
        if _temp != 0:
            for i in range(1, 6):
                x = position.X - i * x_vector
                y = position.Y - i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if _temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # blank = 2, quit
                    elif _temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # block by opponent
                    if _temp == 1:
                        sides += 1
                    elif _temp == 2:
                        sides1 += 1

        score = 0
        if count == 4:
            score = 10000
        elif count1 == 4:
            score = 9000
        elif count == 3:
            if sides == 0:
                score = 1000
            elif sides == 1:
                score = 100
            else:
                score = 0
        elif count1 == 3:
            if sides1 == 0:
                score = 900
            elif sides1 == 1:
                score = 90
            else:
                score = 0
        elif count == 2:
            if sides == 0:
                score = 1000
            elif sides == 1:
                score = 10
            else:
                score = 0
        elif count1 == 2:
            if sides1 == 0:
                score = 90
            elif sides1 == 1:
                score = 9
            else:
                score = 0
        elif count == 1:
            score = 10
        elif count1 == 1:
            score = 9
        else:
            score = 0

        if blank or blank1:
            score /= 1.1

        return score  
    
    def computer_drop3(self):
        position = None
        score = 0
        for i in range(self._node_numbers):
            for j in range(self._node_numbers):
                if self._board[j][i] == 0:
                    score1 = self._get_position_score3(Position(i, j))
                    if score1 > score:
                        score = score1
                        position = Position(i, j)
                    elif score1 == score and score1 > 0:
                        u = random.randint(0, 1)
                        if u  == 0:
                            position = Position(i, j)
        self._board[position.Y][position.X] = self._computer.Value
        return position

    def _get_position_score3(self, position):
        score = 0
        for v in vector:
            score += self._get_vector_score3(position, v[0], v[1])
        return score

    def _get_vector_score3(self, position, x_vector, y_vector):
        count = 0   # number of computer continuous chess
        count1 = 0  # number of people continuous chess
        blank = None   # blank between continuous chess of computer
        blank1 = None  # blank between continuous chess of people
        sides = 0    # block on continuous chess (two sides) of computer
        sides1 = 0   # block on continuous chess (two sides) of people

        # 1 means computer chess besides, 2 means people chess
        temp = self._get_chess_color(position, x_vector, y_vector, True)
        if temp != 0:
            for i in range(1, 6):
                x = position.X + i * x_vector
                y = position.Y + i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # blank = 2, quit
                    elif temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # block by opponent
                    if temp == 1:
                        sides += 1
                    elif temp == 2:
                        sides1 += 1

        if blank is False:
            blank = None
        if blank1 is False:
            blank1 = None

        _temp = self._get_chess_color(position, -x_vector, -y_vector, True)
        if _temp != 0:
            for i in range(1, 6):
                x = position.X - i * x_vector
                y = position.Y - i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if _temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # blank = 2, quit
                    elif _temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # block by opponent
                    if _temp == 1:
                        sides += 1
                    elif _temp == 2:
                        sides1 += 1

        score = 0
        if count == 4:
            score = 10000
        elif count1 == 4:
            score = 9000
        elif count == 3:
            if sides == 0:
                score = 1000
            elif sides == 1:
                score = 100
            else:
                score = 0
        elif count1 == 3:
            if sides1 == 0:
                score = 900
            elif sides1 == 1:
                score = 90
            else:
                score = 0
        elif count == 2:
            if sides == 0:
                score = 100
            elif sides == 1:
                score = 10
            else:
                score = 0
        elif count1 == 2:
            if sides1 == 0:
                score = 90
            elif sides1 == 1:
                score = 9
            else:
                score = 0
        elif count == 1:
            score = 10
        elif count1 == 1:
            score = 9
        else:
            score = 0

        if blank or blank1:
            score /= 2

        return score  

    # determin the chess color besides
    def _get_chess_color(self, position, x_vector, y_vector, next):
        x = position.X + x_vector
        y = position.Y + y_vector
        if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
            if self._board[y][x] == self._computer.Value:
                return 1
            elif self._board[y][x] == self._people.Value:
                return 2
            else:
                if next:
                    return self._get_chess_color(Position(x, y), x_vector, y_vector, False)
                else:
                    return 0
        else:
            return 0

        

game_intro()
pygame.quit()
quit() 
```
