```python
import sys
import random
import pygame
from pygame.locals import *
import pygame.gfxdraw
from checkerboard import Check_all, Black, White, vector, Position


Length = 30  # 棋盘每个点时间的间隔
Node_Numbers = 19  # 棋盘每行/每列点数
Outer_Width = 20  # 棋盘外宽度
Border_Width = 4  # 边框宽度
Inside_Width = 4  # 边框跟实际的棋盘之间的间隔
Border_Length = Length * (Node_Numbers - 1) + Inside_Width * 2 + Border_Width  # 边框线的长度
Start_X = Start_Y = Outer_Width + int(Border_Width / 2) + Inside_Width  # 网格线起点（左上角）坐标
SCREEN_HEIGHT = Length * (Node_Numbers - 1) + Outer_Width * 2 + Border_Width + Inside_Width * 2  # 游戏屏幕的高
SCREEN_WIDTH = SCREEN_HEIGHT + 200  # 游戏屏幕的宽

Radius = Length // 2 - 3  # 棋子半径
Radius2 = Length // 2 + 3
Board_Color = (0xE3, 0x92, 0x65)
#Board_Color = (0xE3, 0x92, 0x65)  # 棋盘颜色
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (200, 30, 30)
BLUE = (30, 30, 200)

RIGHT_INFO_POS_X = SCREEN_HEIGHT + Radius2 * 2 + 10

clock = pygame.time.Clock()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
def game_intro():
    pygame.init()
    intro = True
    screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    while intro:
        for event in pygame.event.get():
#             print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
        screen.fill(WHITE)
        button("GO", 200, 100, 30, 30, RED, BLUE,d1)
        button("GO", 350, 100, 30, 30, RED, BLUE,d2)
        button("GO", 500, 100, 30, 30, RED, BLUE,d3)
        pygame.display.update()
        clock.tick(15)
    
def _text(screen, font, x, y, text, color=(255, 255, 255)):
    imgText = font.render(text, True, color)
    screen.blit(imgText, (x, y))    
    
def d1():
#     pygame.init()
#     screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    pygame.display.set_caption('五子棋')

    font1 = pygame.font.SysFont('SimHei', 32)
    font2 = pygame.font.SysFont('SimHei', 72)
    fwidth, fheight = font2.size('Black Win')

    checkerboard = Check_all(Node_Numbers)
    current_runner = Black
    winner = None
    com = computer_step(Node_Numbers, White)

    black_win_count = 0
    white_win_count = 0

#     pygame.draw.rect(screen, BLUE, (700,300,30,30))
#     button("GO", 700, 300, 30, 30, RED, BLUE,dd)


    while True :
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                quit()
            elif event.type == KEYDOWN:
                if event.key == K_RETURN:
                    if winner is not None:
                        winner = None
                        current_runner = Black
                        checkerboard = Check_all(Node_Numbers)
                        com = computer_step(Node_Numbers, White)
            elif event.type == MOUSEBUTTONDOWN:
                if winner is None:
                    pressed_array = pygame.mouse.get_pressed()
                    if pressed_array[0]:
                        mouse_pos = pygame.mouse.get_pos()
                        click_point = _get_clickposition(mouse_pos)
                        if click_point is not None:
                            if checkerboard.able_to_drop(click_point):
                                winner = checkerboard.step(current_runner, click_point)
                                if winner is None:
                                    current_runner = _get_next(current_runner)
                                    com.get_people_drop(click_point)
                                    computer_step_point = com.computer_drop1()
                                    winner = checkerboard.step(current_runner, computer_step_point)
                                    if winner is not None:
                                        white_win_count += 1
                                    current_runner = _get_next(current_runner)
                                else:
                                    black_win_count += 1
                        else:
                            print('超出棋盘区域')

        # 画棋盘
        _draw_board(screen)

        # 画棋盘上已有的棋子
        for i, row in enumerate(checkerboard.checkerboard):
            for j, cell in enumerate(row):
                if cell == Black.Value:
                    _draw_chess(screen, Position(j, i), Black.Color)
                elif cell == White.Value:
                    _draw_chess(screen, Position(j, i), White.Color)

        _draw_left_info(screen, font1, current_runner, black_win_count, white_win_count)
#         pygame.draw.rect(screen, BLUE, (700,300,30,30))
#         button("GO", 700, 300, 30, 30, RED, BLUE,dd)
        if winner:
            _text(screen, font2, (SCREEN_WIDTH - fwidth)//2, (SCREEN_HEIGHT - fheight)//2, winner.Name + 'win', RED)

        pygame.display.flip()
def d2():
#     pygame.init()
#     screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    pygame.display.set_caption('五子棋')

    font1 = pygame.font.SysFont('SimHei', 32)
    font2 = pygame.font.SysFont('SimHei', 72)
    fwidth, fheight = font2.size('Black Win')

    checkerboard = Check_all(Node_Numbers)
    current_runner = Black
    winner = None
    com = computer_step(Node_Numbers, White)

    black_win_count = 0
    white_win_count = 0

#     pygame.draw.rect(screen, BLUE, (700,300,30,30))
#     button("GO", 700, 300, 30, 30, RED, BLUE,dd)


    while True :
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                quit()
            elif event.type == KEYDOWN:
                if event.key == K_RETURN:
                    if winner is not None:
                        winner = None
                        current_runner = Black
                        checkerboard = Check_all(Node_Numbers)
                        com = computer_step(Node_Numbers, White)
            elif event.type == MOUSEBUTTONDOWN:
                if winner is None:
                    pressed_array = pygame.mouse.get_pressed()
                    if pressed_array[0]:
                        mouse_pos = pygame.mouse.get_pos()
                        click_point = _get_clickposition(mouse_pos)
                        if click_point is not None:
                            if checkerboard.able_to_drop(click_point):
                                winner = checkerboard.step(current_runner, click_point)
                                if winner is None:
                                    current_runner = _get_next(current_runner)
                                    com.get_people_drop(click_point)
                                    computer_step_point = com.computer_drop2()
                                    winner = checkerboard.step(current_runner, computer_step_point)
                                    if winner is not None:
                                        white_win_count += 1
                                    current_runner = _get_next(current_runner)
                                else:
                                    black_win_count += 1
                        else:
                            print('超出棋盘区域')

        # 画棋盘
        _draw_board(screen)

        # 画棋盘上已有的棋子
        for i, row in enumerate(checkerboard.checkerboard):
            for j, cell in enumerate(row):
                if cell == Black.Value:
                    _draw_chess(screen, Position(j, i), Black.Color)
                elif cell == White.Value:
                    _draw_chess(screen, Position(j, i), White.Color)

        _draw_left_info(screen, font1, current_runner, black_win_count, white_win_count)
#         pygame.draw.rect(screen, BLUE, (700,300,30,30))
#         button("GO", 700, 300, 30, 30, RED, BLUE,dd)
        if winner:
            _text(screen, font2, (SCREEN_WIDTH - fwidth)//2, (SCREEN_HEIGHT - fheight)//2, winner.Name + 'win', RED)

        pygame.display.flip()
        
def d3():
#     pygame.init()
#     screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    pygame.display.set_caption('五子棋')

    font1 = pygame.font.SysFont('SimHei', 32)
    font2 = pygame.font.SysFont('SimHei', 72)
    fwidth, fheight = font2.size('Black Win')

    checkerboard = Check_all(Node_Numbers)
    current_runner = Black
    winner = None
    com = computer_step(Node_Numbers, White)

    black_win_count = 0
    white_win_count = 0

#     pygame.draw.rect(screen, BLUE, (700,300,30,30))
#     button("GO", 700, 300, 30, 30, RED, BLUE,dd)


    while True :
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                quit()
            elif event.type == KEYDOWN:
                if event.key == K_RETURN:
                    if winner is not None:
                        winner = None
                        current_runner = Black
                        checkerboard = Check_all(Node_Numbers)
                        com = computer_step(Node_Numbers, White)
            elif event.type == MOUSEBUTTONDOWN:
                if winner is None:
                    pressed_array = pygame.mouse.get_pressed()
                    if pressed_array[0]:
                        mouse_pos = pygame.mouse.get_pos()
                        click_point = _get_clickposition(mouse_pos)
                        if click_point is not None:
                            if checkerboard.able_to_drop(click_point):
                                winner = checkerboard.step(current_runner, click_point)
                                if winner is None:
                                    current_runner = _get_next(current_runner)
                                    com.get_people_drop(click_point)
                                    computer_step_point = com.computer_drop3()
                                    winner = checkerboard.step(current_runner, computer_step_point)
                                    if winner is not None:
                                        white_win_count += 1
                                    current_runner = _get_next(current_runner)
                                else:
                                    black_win_count += 1
                        else:
                            print('超出棋盘区域')

        # 画棋盘
        _draw_board(screen)

        # 画棋盘上已有的棋子
        for i, row in enumerate(checkerboard.checkerboard):
            for j, cell in enumerate(row):
                if cell == Black.Value:
                    _draw_chess(screen, Position(j, i), Black.Color)
                elif cell == White.Value:
                    _draw_chess(screen, Position(j, i), White.Color)

        _draw_left_info(screen, font1, current_runner, black_win_count, white_win_count)
#         pygame.draw.rect(screen, BLUE, (700,300,30,30))
#         button("GO", 700, 300, 30, 30, RED, BLUE,dd)
        if winner:
            _text(screen, font2, (SCREEN_WIDTH - fwidth)//2, (SCREEN_HEIGHT - fheight)//2, winner.Name + 'win', RED)

        pygame.display.flip()
        
def button (msg, x, y, w, h, ic, ac, action=None):

    mouse =pygame.mouse.get_pos()
    click = pygame.mouse.get_pressed()
#     print(click)
    if x + w > mouse[0] > x and y + h > mouse[1] > y:
        pygame.draw.rect(screen, ac, (x,y,w,h))
        if click[0] == 1 and action != None:
            action()
    else:
        pygame.draw.rect(screen, ic, (x,y,w,h))
        
def _get_next(current_runner):
    if current_runner == Black:
        return White
    else:
        return Black
    
# 画棋盘
def _draw_board(screen):
    # 填充棋盘背景色
    screen.fill(Board_Color)
    # 画棋盘网格线外的边框
    pygame.draw.rect(screen, BLACK, (Outer_Width, Outer_Width, Border_Length, Border_Length), Border_Width)
    # 画网格线
    for i in range(Node_Numbers):
        pygame.draw.line(screen, BLACK,
                         (Start_Y, Start_Y + Length * i),
                         (Start_Y + Length * (Node_Numbers - 1), Start_Y + Length * i),
                         1)
    for j in range(Node_Numbers):
        pygame.draw.line(screen, BLACK,
                         (Start_X + Length * j, Start_X),
                         (Start_X + Length * j, Start_X + Length * (Node_Numbers - 1)),
                         1)
#     画星位和天元
    for i in (3, 9, 15):
        for j in (3, 9, 15):
            if i == j == 9:
                radius = 5
            else:
                radius = 3
            # pygame.draw.circle(screen, BLACK, (Start_X + Length * i, Start_Y + Length * j), radius)
            pygame.gfxdraw.aacircle(screen, Start_X + Length * i, Start_Y + Length * j, radius, BLACK)
            pygame.gfxdraw.filled_circle(screen, Start_X + Length * i, Start_Y + Length * j, radius, BLACK)
            
# 画棋子
def _draw_chess(screen, position, chess_color):
    # pygame.draw.circle(screen, chess_color, (Start_X + Length * position.X, Start_Y + Length * position.Y), Stone_Radius)
    pygame.gfxdraw.aacircle(screen, Start_X + Length * position.X, Start_Y + Length * position.Y, Radius, chess_color)
    pygame.gfxdraw.filled_circle(screen, Start_X + Length * position.X, Start_Y + Length * position.Y, Radius, chess_color)
    
# 画左侧信息显示
def _draw_left_info(screen, font, current_runner, black_win_count, white_win_count):
    _draw_chess_pos(screen, (SCREEN_HEIGHT + Radius2, Start_X + Radius2), Black.Color)
    _draw_chess_pos(screen, (SCREEN_HEIGHT + Radius2, Start_X + Radius2 * 4), White.Color)

    _text(screen, font, RIGHT_INFO_POS_X, Start_X + 3, 'player', BLUE)
    _text(screen, font, RIGHT_INFO_POS_X, Start_X + Radius2 * 3 + 3, 'computer', BLUE)

    _text(screen, font, SCREEN_HEIGHT, SCREEN_HEIGHT - Radius2 * 8, 'history：', BLUE)
    _draw_chess_pos(screen, (SCREEN_HEIGHT + Radius2, SCREEN_HEIGHT - int(Radius2 * 4.5)), Black.Color)
    _draw_chess_pos(screen, (SCREEN_HEIGHT + Radius2, SCREEN_HEIGHT - Radius2 * 2), White.Color)
    _text(screen, font, RIGHT_INFO_POS_X, SCREEN_HEIGHT - int(Radius2 * 5.5) + 3, f'{black_win_count} win', BLUE)
    _text(screen, font, RIGHT_INFO_POS_X, SCREEN_HEIGHT - Radius2 * 3 + 3, f'{white_win_count} win', BLUE)
    

def _draw_chess_pos(screen, pos, chess_color):
    pygame.gfxdraw.aacircle(screen, pos[0], pos[1], Radius2, chess_color)
    pygame.gfxdraw.filled_circle(screen, pos[0], pos[1], Radius2, chess_color)
    
# 根据鼠标点击位置，返回游戏区坐标
def _get_clickposition(click_pos):
    pos_x = click_pos[0] - Start_X
    pos_y = click_pos[1] - Start_Y
    if pos_x < -Inside_Width or pos_y < -Inside_Width:
        return None
    x = pos_x // Length
    y = pos_y // Length
    if pos_x % Length > Radius:
        x += 1
    if pos_y % Length > Radius:
        y += 1
    if x >= Node_Numbers or y >= Node_Numbers:
        return None

    return Position(x, y)


class computer_step:
    def __init__(self, node_numbers, chess):
        self._node_numbers = node_numbers
        self._computer = chess
        if chess == White:
            self._people = Black
        else:
            self._people = White
        self._board = [[0] * node_numbers for _ in range(node_numbers)]

    def get_people_drop(self, position):
        self._board[position.Y][position.X] = self._people.Value

    def computer_drop1(self):
        position = None
        score = 0
        for i in range(self._node_numbers):
            for j in range(self._node_numbers):
                if self._board[j][i] == 0:
                    score1 = self._get_position_score1(Position(i, j))
                    if score1 > score:
                        score = score1
                        position = Position(i, j)
                    elif score1 == score and score1 > 0:
                        u = random.randint(0, 1)
                        if u  == 0:
                            position = Position(i, j)
        self._board[position.Y][position.X] = self._computer.Value
        return position

    def _get_position_score1(self, position):
        score = 0
        for v in vector:
            score += self._get_vector_score1(position, v[0], v[1])
        return score

    def _get_vector_score1(self, position, x_vector, y_vector):
        count = 0   # 落子处我方连续子数
        count1 = 0  # 落子处对方连续子数
        blank = None   # 我方连续子中有无空格
        blank1 = None  # 对方连续子中有无空格
        sides = 0    # 我方连续子两端有无阻挡
        sides1 = 0   # 对方连续子两端有无阻挡

        # 如果是 1 表示是边上是我方子，2 表示敌方子
        temp = self._get_chess_color(position, x_vector, y_vector, True)
        if temp != 0:
            for i in range(1, 6):
                x = position.X + i * x_vector
                y = position.Y + i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # 遇到第二个空格退出
                    elif temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # 遇到边也就是阻挡
                    if temp == 1:
                        sides += 1
                    elif temp == 2:
                        sides1 += 1

        if blank is False:
            blank = None
        if blank1 is False:
            blank1 = None

        _temp = self._get_chess_color(position, -x_vector, -y_vector, True)
        if _temp != 0:
            for i in range(1, 6):
                x = position.X - i * x_vector
                y = position.Y - i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if _temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # 遇到第二个空格退出
                    elif _temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # 遇到边也就是阻挡
                    if _temp == 1:
                        sides += 1
                    elif _temp == 2:
                        sides1 += 1

        score = 0
        if count == 4:
            score = 10000
        elif count1 == 4:
            score = 9000
        elif count == 3:
            if sides == 0:
                score = 1000
            elif sides == 1:
                score = 100
            else:
                score = 0
        elif count1 == 3:
            if sides1 == 0:
                score = 5
            elif sides1 == 1:
                score = 4
            else:
                score = 0
        elif count == 2:
            if sides == 0:
                score = 100
            elif sides == 1:
                score = 10
            else:
                score = 0
        elif count1 == 2:
            if sides1 == 0:
                score = 3
            elif sides1 == 1:
                score = 2
            else:
                score = 0
        elif count == 1:
            score = 10
        elif count1 == 1:
            score = 1
        else:
            score = 0

        if blank or blank1:
            score /= 2

        return score
    def computer_drop2(self):
        position = None
        score = 0
        for i in range(self._node_numbers):
            for j in range(self._node_numbers):
                if self._board[j][i] == 0:
                    score1 = self._get_position_score2(Position(i, j))
                    if score1 > score:
                        score = score1
                        position = Position(i, j)
                    elif score1 == score and score1 > 0:
                        u = random.randint(0, 1)
                        if u  == 0:
                            position = Position(i, j)
        self._board[position.Y][position.X] = self._computer.Value
        return position

    def _get_position_score2(self, position):
        score = 0
        for v in vector:
            score += self._get_vector_score1(position, v[0], v[1])
        return score

    def _get_vector_score2(self, position, x_vector, y_vector):
        count = 0   # 落子处我方连续子数
        count1 = 0  # 落子处对方连续子数
        blank = None   # 我方连续子中有无空格
        blank1 = None  # 对方连续子中有无空格
        sides = 0    # 我方连续子两端有无阻挡
        sides1 = 0   # 对方连续子两端有无阻挡

        # 如果是 1 表示是边上是我方子，2 表示敌方子
        temp = self._get_chess_color(position, x_vector, y_vector, True)
        if temp != 0:
            for i in range(1, 6):
                x = position.X + i * x_vector
                y = position.Y + i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # 遇到第二个空格退出
                    elif temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # 遇到边也就是阻挡
                    if temp == 1:
                        sides += 1
                    elif temp == 2:
                        sides1 += 1

        if blank is False:
            blank = None
        if blank1 is False:
            blank1 = None

        _temp = self._get_chess_color(position, -x_vector, -y_vector, True)
        if _temp != 0:
            for i in range(1, 6):
                x = position.X - i * x_vector
                y = position.Y - i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if _temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # 遇到第二个空格退出
                    elif _temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # 遇到边也就是阻挡
                    if _temp == 1:
                        sides += 1
                    elif _temp == 2:
                        sides1 += 1

        score = 0
        if count == 4:
            score = 10000
        elif count1 == 4:
            score = 9000
        elif count == 3:
            if sides == 0:
                score = 1000
            elif sides == 1:
                score = 100
            else:
                score = 0
        elif count1 == 3:
            if sides1 == 0:
                score = 5
            elif sides1 == 1:
                score = 4
            else:
                score = 0
        elif count == 2:
            if sides == 0:
                score = 100
            elif sides == 1:
                score = 10
            else:
                score = 0
        elif count1 == 2:
            if sides1 == 0:
                score = 3
            elif sides1 == 1:
                score = 2
            else:
                score = 0
        elif count == 1:
            score = 10
        elif count1 == 1:
            score = 1
        else:
            score = 0

        if blank or blank1:
            score /= 2

        return score
    
    def computer_drop3(self):
        position = None
        score = 0
        for i in range(self._node_numbers):
            for j in range(self._node_numbers):
                if self._board[j][i] == 0:
                    score1 = self._get_position_score3(Position(i, j))
                    if score1 > score:
                        score = score1
                        position = Position(i, j)
                    elif score1 == score and score1 > 0:
                        u = random.randint(0, 1)
                        if u  == 0:
                            position = Position(i, j)
        self._board[position.Y][position.X] = self._computer.Value
        return position

    def _get_position_score3(self, position):
        score = 0
        for v in vector:
            score += self._get_vector_score3(position, v[0], v[1])
        return score

    def _get_vector_score3(self, position, x_vector, y_vector):
        count = 0   # 落子处我方连续子数
        count1 = 0  # 落子处对方连续子数
        blank = None   # 我方连续子中有无空格
        blank1 = None  # 对方连续子中有无空格
        sides = 0    # 我方连续子两端有无阻挡
        sides1 = 0   # 对方连续子两端有无阻挡

        # 如果是 1 表示是边上是我方子，2 表示敌方子
        temp = self._get_chess_color(position, x_vector, y_vector, True)
        if temp != 0:
            for i in range(1, 6):
                x = position.X + i * x_vector
                y = position.Y + i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # 遇到第二个空格退出
                    elif temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # 遇到边也就是阻挡
                    if temp == 1:
                        sides += 1
                    elif temp == 2:
                        sides1 += 1

        if blank is False:
            blank = None
        if blank1 is False:
            blank1 = None

        _temp = self._get_chess_color(position, -x_vector, -y_vector, True)
        if _temp != 0:
            for i in range(1, 6):
                x = position.X - i * x_vector
                y = position.Y - i * y_vector
                if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
                    if _temp == 1:
                        if self._board[y][x] == self._computer.Value:
                            count += 1
                            if blank is False:
                                blank = True
                        elif self._board[y][x] == self._people.Value:
                            sides1 += 1
                            break
                        else:
                            if blank is None:
                                blank = False
                            else:
                                break   # 遇到第二个空格退出
                    elif _temp == 2:
                        if self._board[y][x] == self._computer.Value:
                            sides1 += 1
                            break
                        elif self._board[y][x] == self._people.Value:
                            count1 += 1
                            if blank1 is False:
                                blank1 = True
                        else:
                            if blank1 is None:
                                blank1 = False
                            else:
                                break
                else:
                    # 遇到边也就是阻挡
                    if _temp == 1:
                        sides += 1
                    elif _temp == 2:
                        sides1 += 1

        score = 0
        if count == 4:
            score = 10000
        elif count1 == 4:
            score = 9000
        elif count == 3:
            if sides == 0:
                score = 1000
            elif sides == 1:
                score = 100
            else:
                score = 0
        elif count1 == 3:
            if sides1 == 0:
                score = 5
            elif sides1 == 1:
                score = 4
            else:
                score = 0
        elif count == 2:
            if sides == 0:
                score = 100
            elif sides == 1:
                score = 10
            else:
                score = 0
        elif count1 == 2:
            if sides1 == 0:
                score = 3
            elif sides1 == 1:
                score = 2
            else:
                score = 0
        elif count == 1:
            score = 10
        elif count1 == 1:
            score = 1
        else:
            score = 0

        if blank or blank1:
            score /= 2

        return score  

    # 判断指定位置处在指定方向上是我方子、对方子、空
    def _get_chess_color(self, position, x_vector, y_vector, next):
        x = position.X + x_vector
        y = position.Y + y_vector
        if 0 <= x < self._node_numbers and 0 <= y < self._node_numbers:
            if self._board[y][x] == self._computer.Value:
                return 1
            elif self._board[y][x] == self._people.Value:
                return 2
            else:
                if next:
                    return self._get_chess_color(Position(x, y), x_vector, y_vector, False)
                else:
                    return 0
        else:
            return 0


# if __name__ == '__main__':
#     main()
game_intro()
pygame.quit()
quit()    
```
